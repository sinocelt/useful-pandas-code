```python
import pandas as pd
```

# Special notes:
By default, sorting will be ascending. There are flags you can include to change this. This notebook will have many examples of sorting.

```python
# read in csv and turn it into a dataframe
df = pd.read_csv('salary.csv', delimiter='\t')
```

```python
df
```

```python
df.columns
```

# Sort dataframe by a single column

## sort data by 'name' column ascending

```python
df.sort_values(by='name')
```

## now using a variable

```python
column_to_sort_by = 'name'
```

```python
df.sort_values(by=column_to_sort_by)
```

## sort data by 'name' column descending

```python
df.sort_values(by='name', ascending = False)
```

# sort data by 'yearly_salary' column ascending

```python
# note that ascending = True is not necessary to have here (it is the default). But I show it just to be educational
df.sort_values(by='yearly_salary', ascending = True)
```

# sort data by 'yearly_salary' column descending

```python
df.sort_values(by='yearly_salary', ascending = False)
```

# sort by 2 or more columns

# *** NOTE it will actually do lexicographic sorting order

```python
# sort career column so it is easier to sort later. Lowercase the data
df['career'] = df['career'].apply(lambda x: x.lower())
```

```python
df['career']
```

# Sort by career ascending and then by name ascending (you can change the ascending flag if you want it differently

```python
df.sort_values(by=['career', 'name'], ascending = [True, True])
```
