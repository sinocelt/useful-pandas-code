```python
# datagoes from 1960 to 2024. Make a list of those numbers
years_of_data = list(range(1960, 2025))
```

```python
# years_of_data
```

```python
# convert the list into strings so I can iterate over column names
years_of_data_str = [str(x) for x in years_of_data]
```

```python
# years_of_data_str
```

```python
column_names = gdp_growth_df_top_2024.columns
```

```python
# USE TOP 5 or top 10
```

```python
column_names = gdp_growth_df_top_2024.columns

dataframes_list = []

# for index, row in gdp_growth_df_top_2024.iloc[0:1].iterrows():
# top_10_gdp_growth_countries_2024_df
# for index, row in gdp_growth_df_top_2024.iterrows():
# for index, row in top_10_gdp_growth_countries_2024_df.iterrows():
for index, row in top_5_gdp_growth_countries_2024_df.iterrows():

    cur_country_name = row['Country Name']

    country_list = []
    year_list = []
    cur_gdp_growth_rate_list = []

    for year in years_of_data_str:
        # add country and year
        country_list.append(cur_country_name)
        year_list.append(year)
        cur_gdp_growth_rate_list.append(row[year])

    # make dataframe of current row to later combine into a larger dataframe
    cur_data_dict = {
        'country': country_list,
        'year': year_list,
        'gdp_growth_rate': cur_gdp_growth_rate_list
    }
    cur_df = pd.DataFrame(data=cur_data_dict)
    
    dataframes_list.append(cur_df)
```

```python
top_5_countries_combined_df = pd.concat(dataframes_list)
```

```python
top_5_countries_combined_df
```
